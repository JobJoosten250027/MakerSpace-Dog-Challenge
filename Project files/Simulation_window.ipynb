```python
#Importing necessary parts

import mujoco
import mujoco_viewer
import numpy as np
import threading

print(f'Mujoco Version: {mujoco.mj_version}')
print(f'Numpy version: {np.version}')
```

```python
#Loading of the Unitree GO2-W model

model_path = '../unitree_mujoco/unitree_robots/go2w/go2w.xml'  # Path to the go2_w file
model = mujoco.MjModel.from_xml_path(model_path)
data = mujoco.MjData(model)

print(f"Model loaded successfully!")
print(f"Number of actuators: {model.nu}")
print(f"Number of joints: {model.njnt}")
```

```python
def run_go2w_simulation(model, data, num_steps=5000):
    """
    Run Go2-W simulation in a separate window
    """
    # Create viewer (opens separate window)
    viewer = mujoco_viewer.MujocoViewer(model, data)
    
    try:
        for step in range(num_steps):
            # Basic wheel control example
            # Set all wheel velocities (adjust indices based on your model)
            # Typically: [FL_wheel, FR_wheel, RL_wheel, RR_wheel]
            
            # Example: Move forward
            data.ctrl[:] = 0  # Reset all controls
            # data.ctrl[0:4] = [5.0, 5.0, 5.0, 5.0]  # All wheels forward
            
            # Step the simulation
            mujoco.mj_step(model, data)
            
            # Update viewer
            viewer.render()
            
    except KeyboardInterrupt:
        print("Simulation stopped by user")
    finally:
        viewer.close()
    
    print("Simulation completed!")

# Run it
run_go2w_simulation(model, data, num_steps=3000)
```
